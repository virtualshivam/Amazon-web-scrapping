```python
def check_price():
    #Importing Libraries

    from bs4 import BeautifulSoup
    import requests
    import time
    import datetime
    import csv
    import smtplib
    
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36", "Accept-Encoding": "gzip, deflate, br, zstd", "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7"}
    
    URL = r"https://www.amazon.in/CreativiT-Graphic-Printed-Graphics-D00255-50_White_Medium/dp/B08NWSH4NB/ref=sr_1_8?crid=QU9Z1SCAHKWW&dib=eyJ2IjoiMSJ9.BvCe4QmEZgftrs1aQ-N_sM3jR3IUhaOgKRAFidb3VTUnla9bVmi6gk6xGIWjH6DyNMiAQthrEKipDd1wQaqy_-HrpFW7yWh8yA8oqNh65xrLCqWBF9gQP2xA4YilTzodt2YCafy0YtcjRGaDOIkQaWjwxTMlZV266WJCBPLjBCVN4YrljjVm1w2dvP-5Ro85Qa0A2pt8D_GroBygLThKazGbHfEIOe8BTNXZ09dtcrUN-n7ibnH-ppE1fzO3qf_KVNAU43Koj0lb3Jl72JmosUkcDNMcL89kXmldliPgYms.3fpNzFt87Tfpvbi8AZRnxU1xrx_YVV1WR85160DalO0&dib_tag=se&keywords=Funny+Got+Data+MIS+Data+Systems+Business+Analyst+T-Shirt&nsdOptOutParam=true&qid=1720898956&sprefix=funny+got+data+mis+data+systems+business+analyst+t-shirt%2Caps%2C327&sr=8-8"
    page =requests.get(URL, headers = header)
    soup1 = BeautifulSoup(page.content,"html")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    title = soup2.find("span",id="productTitle").get_text().strip()
    price = soup2.find(class_="a-price-whole").text.strip()
    data_collection_datetime = datetime.datetime.now()
    row_data = [data_collection_datetime,title,price]
    with open(r"C:\Users\shivam\Desktop\Master\Python\AmazonWebScrapping\AmazonWebScrappingDataSet.csv","a+",newline="",encoding="UTF8") as f:
        writer = csv.writer(f)
        writer.writerow(row_data)
```
